```python
import pandas as pd
import numpy as np
from scipy import stats #for box-cox transformation
from mlxtend.preprocessing import minmax_scaling 
import seaborn as sns # for plot
import matplotlib.pyplot as plt #for plot

np.random.seed(0)
```

# scaling

why: when you want to compare 2 variables on equal scale, eg height and weight, if doesn't scale, 1 Yen and 1 dollar would be the same importance  
when: support vector machine, k nearest neighbors  
how: min-max scaling  

```python
#sample data
data=np.random.exponential(size=1000)
scaled_data=minmax_scaling(data,columns=[0]) #scale the first col between 0 to 1
```

```python
fig,ax=plt.subplots(1,2)
sns.distplot(data,ax=ax[0])
ax[0].set_title('original data')
sns.distplot(scaled_data,ax=ax[1])
ax[1].set_title('scaled data')
```

```python
fig #shape doesn't change, range change from 0-10 to 0-1
```

# normalization

what: change observation to normal distribution  
why: some techniques has normal distribution assumption  
(t-test, ANOVA, linear regression, LDA, gaussian naive bayes)  
(any method with 'Gaussian' is name probably assumes normality)  
how: box-cox  

```python
normalized_data=stats.boxcox(data)
```

```python
fig,ax=plt.subplots(1,2)
sns.distplot(data,ax=ax[0])
ax[0].set_title('original data')
sns.distplot(normalized_data[0],ax=ax[1])
ax[1].set_title('normalized data')
```

```python
fig #shape change
```
