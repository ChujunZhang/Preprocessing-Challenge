# sklearn imputation

```python
subset=nfl_data.loc[:,'SideofField':'PlayType']
subset.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#define to score MAD
def num_MAD(X_train,X_test,Y_train,Y_test):
    model=RandomForestRegressor()
    model.fit(X_train,Y_train)
    preds=model.predict(X_test)
    return mean_absolute_error(Y_test,preds)
```

```python
#only impute numeric variables
num_subset=subset.select_dtypes(exclude=['object'])
nfl_data.EPA=nfl_data.EPA.fillna(0)
#divide training and test
X_train,X_test,Y_train,Y_test=train_test_split(num_subset,nfl_data.EPA, \
                                              train_size=0.7,test_size=0.3)
```

```python
#what if drop all missing columns
col_missing=[col for col in X_train.columns if X_train[col].isnull().any()]
reduce_X_train=X_train.drop(col_missing,axis=1)
reduce_X_test=X_test.drop(col_missing,axis=1)
```

```python
MAD=num_MAD(reduce_X_train,reduce_X_test,Y_train,Y_test)
print("MAD when removing all missing columns is: %.10f" %MAD)
```

```python
#impute missing values
from sklearn.preprocessing import Imputer
```

```python
my_imputer = Imputer() #default strategy is mean
imputed_X_train = my_imputer.fit_transform(X_train)
imputed_X_test = my_imputer.fit_transform(X_test)
```

```python
MAD2=num_MAD(imputed_X_train,imputed_X_test,Y_train,Y_test)
print("MAD when imputing all missing columns is: %.10f" %MAD)
```

```python
#label which is missing and also impute them
imputed_X_train_plus=X_train.copy()
imputed_X_test_plus=X_test.copy()
cols_with_missing = (col for col in X_train.columns 
                                 if X_train[col].isnull().any())
for col in cols_with_missing:
    imputed_X_train_plus[col + '_was_missing'] = imputed_X_train_plus[col].isnull()
    imputed_X_test_plus[col + '_was_missing'] = imputed_X_test_plus[col].isnull()
    
my_imputer = Imputer()
imputed_X_train_plus = my_imputer.fit_transform(imputed_X_train_plus)
imputed_X_test_plus = my_imputer.fit_transform(imputed_X_test_plus)
```

```python
MAD3=num_MAD(imputed_X_train_plus,imputed_X_test_plus,Y_train,Y_test)
print("MAD when imputing all missing columns and showing which is imputed is: %.10f" %MAD)
```
